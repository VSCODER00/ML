```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Travel.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["Gender"].value_counts()
```

```python
df['Gender']=df['Gender'].replace('Fe Male','Female')
```

```python
df['MaritalStatus'].value_counts()
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace('Single', 'Unmarried')
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df['TypeofContact'].fillna(df['TypeofContact'].mode(),inplace=True)
```

```python
df.DurationOfPitch.fillna(df.DurationOfPitch.median(), inplace=True)
```

```python
df.NumberOfFollowups.fillna(df.NumberOfFollowups.mode()[0], inplace=True)
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode()[0], inplace=True)
df.NumberOfTrips.fillna(df.NumberOfTrips.median(), inplace=True)
df.NumberOfChildrenVisiting.fillna(df.NumberOfChildrenVisiting.mode()[0], inplace=True)
df.MonthlyIncome.fillna(df.MonthlyIncome.median(), inplace=True)
```

```python
df.TypeofContact.fillna(df.TypeofContact.mode()[0], inplace=True)
```

```python
df.drop('CustomerID',axis=1,inplace=True)
```

```python
df['TotalVisiting'] = df['NumberOfPersonVisiting'] + df['NumberOfChildrenVisiting']
df.drop(columns=['NumberOfPersonVisiting', 'NumberOfChildrenVisiting'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.to_csv('CleanedTravel.csv')
```

```python
x=df.drop('ProdTaken',axis=1)
y=df['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
df.dtypes
```

```python
cat_features = x.select_dtypes(include="object").columns
print(cat_features)
```

```python
df['Designation'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(drop='first')
encoded=encoder.fit_transform(x_train[cat_features]).toarray()
one_hot_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(cat_features),index=x_train.index)
x_train=pd.concat([x_train,one_hot_df],axis=1).drop(columns=cat_features)
```

```python

encoded=encoder.transform(x_test[cat_features]).toarray()
one_hot_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(cat_features),index=x_test.index)
x_test=pd.concat([x_test,one_hot_df],axis=1).drop(columns=cat_features)
```

Here, we describe an efficient way to call various models:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
models={
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}
for i in models:
    model=models[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    clreport=classification_report(y_test,y_pred)
    print("The classification report for ", i ,"\n",clreport)
```

Hyperparameter tuning:

```python
rf_params = {"max_depth": [5, 8, 15, None, 10],
             "max_features": [5, 7, "auto", 8],
             "min_samples_split": [2, 8, 15, 20],
             "n_estimators": [100, 200, 500, 1000]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
grid=RandomizedSearchCV(model,param_distributions=rf_params,cv=5)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
model2=RandomForestClassifier(n_estimators=200,min_samples_split=2,max_features=7,max_depth=15)
```

```python
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid1=GridSearchCV(model,param_grid=rf_params,cv=5,n_jobs=-1)
grid1.fit(x_train,y_train)
```
