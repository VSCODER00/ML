```python
import numpy as np
import pandas as pd
```

```python
with open('dataset_book.txt') as f:
    lines = f.read().lower()
```

```python
type(lines)
```

```python
lines
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()
tokenizer.fit_on_texts([lines])
len(tokenizer.word_index)
```

```python
import pickle
with open("tokenizer.pkl", "wb") as f:
    pickle.dump(tokenizer, f)
```

```python
tokenizer.word_index
```

```python
sentences=[]
for line in lines.split('\n'):
    if(line==''):
        continue
    vectorized=tokenizer.texts_to_sequences([line])[0]
    for j in range(1, len(vectorized)):
        n_gram_sequence = vectorized[:j+1]
        sentences.append(n_gram_sequence)
sentences
```

```python
from tensorflow.keras.utils import pad_sequences
padded = pad_sequences(sentences, padding='pre', value=0)
```

```python
len(padded[324])
```

```python
x,y=padded[:,:-1],padded[:,-1]
```

```python
len(y)
```

```python
num_words=len(tokenizer.word_index)+1
num_words
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense,Embedding
model=Sequential()
model.add(Embedding(input_dim=num_words,output_dim=64))
model.add(SimpleRNN(64))
model.add(Dense(num_words,activation="softmax"))
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
```

```python
history=model.fit(x,y,epochs=15,verbose=1,callbacks=[early_stopping])
```

```python
model.save("SimpleRNN.h5")
```

```python
def predictionPipeline(sentence,next_words):
    sentence=sentence.lower()
    for i in range(next_words):
        token_list=tokenizer.texts_to_sequences([sentence])[0]
        token_list=pad_sequences([token_list],padding='pre',maxlen=17)
        predicted_probabilites=model.predict(token_list)
        predicted_word=tokenizer.index_word[np.argmax(predicted_probabilites)]
        sentence+=' '+predicted_word
    return sentence
```

```python
predictionPipeline("Mathematics is the study of",5)
```

```python
predictionPipeline("I don't know why you",5)
```

```python
predictionPipeline("Once upon a time there",5)
```

Hyperparameter tuning:

```python
import keras_tuner
```

```python
def build_model(hp):
    model = keras.Sequential()
    model.add(Embedding(input_dim=num_words,output_dim=hp.Choice('embedding_dim', [32, 64, 128])))
    model.add(SimpleRNN(units=hp.Choice('rnn_units', [32, 64, 128])))
    model.add(Dense(num_words, activation='softmax'))
    model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
    return model
```

```python
import keras
from kerastuner import HyperModel, RandomSearch

tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=1, 
    directory='rnn_tuning',
    project_name='next_word_prediction')
```

```python
early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
tuner.search(x, y, epochs=20, validation_split=0.1,callbacks=[early_stopping])
```

```python
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best units: {best_hp.get('rnn_units')}")
print(f"Best embedding dim: {best_hp.get('embedding_dim')}")
```

```python
best_model = tuner.hypermodel.build(best_hp)
history = best_model.fit(
    x, y,
    epochs=20,
    validation_split=0.1,
    callbacks=[early_stopping]
)
```

```python
best_model.save("BestRNNModel.h5")
```
