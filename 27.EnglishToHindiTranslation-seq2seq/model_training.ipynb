```python
embedding_dim = 128
hidden_units = 256
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
from tensorflow.keras.models import Model
def build_model():
    # Encoder
    encoder_input = Input(shape = (30, ))        #30 denotes length of each sentence after padding
    encoder_embedding = Embedding(30000, embedding_dim)(encoder_input)
    encoder_lstm = LSTM(hidden_units, return_state=True)
    encoder_outputs, encoder_hidden, encoder_cell = encoder_lstm(encoder_embedding)
    encoder_states = [encoder_hidden, encoder_cell]


    # Decoder
    decoder_input = Input(shape = (39, ))
    decoder_embedding = Embedding(30000, embedding_dim)(decoder_input)
    decoder_lstm = LSTM(hidden_units, return_sequences=True)
    decoder_output = decoder_lstm(decoder_embedding, initial_state = encoder_states)
    decoder_dense = Dense(30000, activation='softmax')
    decoder_output = decoder_dense(decoder_output)

    
    # final model
    model = Model([encoder_input, decoder_input], decoder_output)
    return model
```

```python
model = build_model()
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit([sequences, decoder_input], decoder_target, epochs = 10,batch_size=16)
```

```python
decoder_embedding = seq2seqModel.get_layer("embedding_1")
decoder_dense = seq2seqModel.get_layer("dense")
trained_decoder_lstm = seq2seqModel.get_layer("lstm_1")
hidden_units = trained_decoder_lstm.units
```

```python

dummy_input = Input(shape=(1, embedding_dim))
dummy_state_h = Input(shape=(hidden_units,))
dummy_state_c = Input(shape=(hidden_units,))

decoder_lstm_inf = LSTM(
    hidden_units,
    return_sequences=True,
    return_state=True
)
decoder_lstm_inf(
    dummy_input,
    initial_state=[dummy_state_h, dummy_state_c]
)

# 2. NOW copy weights
decoder_lstm_inf.set_weights(trained_decoder_lstm.get_weights())
```

```python
decoder_state_input_h = Input(shape=(hidden_units,))
decoder_state_input_c = Input(shape=(hidden_units,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]
```

```python
decoder_single_input = Input(shape=(1,))
decoder_embedded = decoder_embedding(decoder_single_input)
```

```python
decoder_outputs, state_h, state_c = decoder_lstm_inf(
    decoder_embedded,
    initial_state=decoder_states_inputs
)
decoder_states = [state_h, state_c]
```

```python
decoder_outputs = decoder_dense(decoder_outputs)
```

```python
decoder_model = Model(
    [decoder_single_input] + decoder_states_inputs,
    [decoder_outputs] + decoder_states
)
```

```python
decoder_model.save("decoder_inference.h5")
```

```python
def decode_sequence(input_seq):
    states = encoder_model.predict(input_seq)

    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tokenizer_hi.word_index['<sos>']

    decoded_sentence = []

    while True:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states
        )

        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        if sampled_token_index == 0:
            break
        sampled_word = tokenizer_hi.index_word[sampled_token_index]

        if sampled_word == '<eos>' or len(decoded_sentence) > 39:
            break

        decoded_sentence.append(sampled_word)

        target_seq[0, 0] = sampled_token_index
        states = [h, c]

    return ' '.join(decoded_sentence)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
def clean_sentence(s):
    s = s.lower().strip()
    s=tokenizer.texts_to_sequences([s])
    s=pad_sequences(s,maxlen=30,padding="post")
    return s
```
